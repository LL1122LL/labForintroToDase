```python
#Q1
x = int(input("请输入一个数字:"))
def IsPrime(x):
    mid = int(x/2) + 1
    for i in range(2,mid+1):
        if x%i==0:
            return False
    return True
flag = IsPrime(x)
if flag:
    print("是质数")
else:
    print("不是质数")
```

```python
#Q3
def insertion_sort(arr):
    for i in range(1, len(arr)):
        current_element = arr[i]
        j = i - 1
        while j >= 0 and current_element < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = current_element

# 示例用法
my_list = [12,9,16,7]
insertion_sort(my_list)
print("排序后的列表:", my_list)
```

直接插入排序是一种简单的排序算法，它通过不断将未排序部分的元素插入到已排序部分中，来达到排序的目的。

流程：
从第一个元素开始，该元素可以认为已经被排序。
取出下一个元素，在已经排序的元素序列中从后向前扫描。
如果该元素（已排序）大于新元素，将该元素移到下一位置。
重复步骤3，直到找到已排序的元素小于或等于新元素的位置。
将新元素插入到该位置。
重复步骤2~5，直到整个数组排序完成。

```python
#希尔排序
def select_sort(array):
    for i in range(len(array)):
        x=i
        for j in range(i,len(array)):
            if array[j] < array[x]:
                x=j
        array[i],array[x] = array[x],array[i]
return array
```

时间复杂度：

最坏情况下：O(n^2)，这是希尔排序的最坏情况时间复杂度。当间隔序列（gap sequence）不适当或者是固定的时候，希尔排序的最坏情况时间复杂度会达到O(n^2)。
平均情况下：取决于所选的间隔序列。平均时间复杂度可以是O(n log^2 n)到O(n^1.5)之间。
希尔排序的性能取决于所选择的间隔序列，不同的间隔序列会导致不同的时间复杂度。通常，希尔排序在大多数情况下都要比普通的插入排序快，但不如一些高级排序算法。

空间复杂度：
希尔排序的空间复杂度是O(1)，因为它仅需要一个额外的变量来进行元素交换，而不需要额外的数据结构来存储数据。
